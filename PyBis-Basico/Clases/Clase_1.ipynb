Listas
- Las listas son estructuras de datos que nos permite almacenar información
- Las listas se definen con corchetes
- Dentro de Python se enumeran los datos empezando con el cero

```python
#            0         1          2          3
armario = ["ropa", "zapatos", "bolsas", "peluches"]

print(armario)
```

```python
print(armario[3])
print(armario[1])
print(armario[0])
```

```python
# Agregar un nuevo elemento a la lista
armario.append("Maleta")

print(armario)

armario.remove("Maleta")

print(armario)
```

Librería Random

- Numpy: enfocado a operaciones numéricas
- pandas: enfocado en manipulación de datos
- scikit-learn: enfocado en algoritmos de machine learning


```python
import random as r
```

```python
# definimos una lista de valores
moneda = ["sol", "águila"]

# Nos permite elegir un elemento al azar
moneda_random = r.choice(moneda)

print(moneda_random)
```

r.randit(a,b): Selecciona un número al azar entre el rango de a y b

r.random(): genera un número aleatorio decimal entre 0 y 1

```python
# Ejemplo 1
print("Seleccionamos un número al azar entre 1 y 27: ", r.randint(1, 27))

print()

# Ejemplo 2
print("Seleccionamos un número al azar entre 0 y 1:", r.random())
```

```python
# Ejemplo 3: Suma de números aleatorios
num1 = r.randint(1, 10)
num2 = r.random()

print("Número 1: ", num1)
print("Número 2: ", num2)

suma = num1 + num2

print("La suma es:", suma)
```

```python
# Podemos redondear los números decimales con la siguiente función
# round(variable, 2)

round(suma, 2)
```

```python
num = round(r.random(), 2)
print(num)
```

Crear variables para representar a un alumno
- Nombre
- Edad
- Promedio
- ¿Está inscrito?

```python
# Zona de definición de lista con la información

listaAlumnos = ["Oscar Ramirez", "Pau Vázquez", "Miguel García", "Juan Reyes"]
listaInscrito = ["Sí", "No"]

# Zona de variables
Edad = r.randint(18, 25)
Promedio = r.randint(70, 100)

# Zona de impresión
print("Nombre: ", r.choice(listaAlumnos))
print("Edad: ", Edad)
print("Promedio: ", Promedio)
print("¿Está inscrito?: ", r.choice(listaInscrito))
```
