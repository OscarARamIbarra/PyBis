```python
# Definir dos variables.
# 1. Calculo en el numerador
# 2. Calculo en el denominador
##############################
# parentesis con numeros --> mult
# - (5 - 1) --> - 1 * (5 - 1)
# + (5 - 1) --> + 1 * (5 - 1)

# Definimos dos variables para los calculos
numerador = -3 * (5 ** 2 -4 *(5 + 10))
denominador = (3 - 1 * (5 - 3)) ** 2

# Efectuamos la division
division = numerador / denominador
print("Resultado de la operación:", division)
```

```python
# len(string): la longitud (el numero de caracteres que contiene) de tu cadena de texto
nombreAlumno = "Oscar Ramirez"
len(nombreAlumno)
```

```python
# Definamos las variables de tipo string
print("Solución inciso a:")
variable1 = "Oscar"
variable2 = "Ramirez"
sumVariables = variable1 + variable2
print("Primera variable:", variable1)
print("Segunda variable:", variable2)
print("Pegamos las variables de tipo string (variable1 + variable2):", sumVariables)

print("-" * 75)

# Calculamos la longitud de cada string
print("Solución inciso b:")
longVariable1 = len(variable1)
longVariable2 = len(variable2)
sumLongVariables = longVariable1 + longVariable2
print("Longitud variable 1=", longVariable1)
print("Longitud variable 2=", longVariable2)
print("long Var1 + long Var2=", sumLongVariables)
```
