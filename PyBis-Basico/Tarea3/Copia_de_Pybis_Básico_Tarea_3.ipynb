**1. Generar fecha aleatoria**

Crea una funciÃ³n llamada fecha_aleatoria(inicio, fin) que reciba dos fechas (en formato "YYYY-MM-DD") y devuelva una fecha aleatoria dentro de ese rango.

Ejemplo de uso:

```python
fecha = fecha_aleatoria("2025-01-01", "2025-01-31")
print(fecha)  # Ejemplo: 2025-01-17
```

**Pistas:**

* Usa pd.date_range() para generar las fechas.
* Usa random.choice() para elegir una.
(2 pts)


````python
##########
# Solucion
##########


# Coloca aca tu solucion ðŸ‘‡
import pandas as pd
import datetime as dt
import random as r

def fecha_aleatoria(inicio, fin):
  fechas = pd.date_range(inicio, fin, freq="1d")
  eleccion = r.choice(fechas)
  return eleccion

fecha = fecha_aleatoria("2025-01-01", "2025-01-31")
print("la fecha aleatoria es: ",fecha)
````

**2. Generar venta aleatoria**

Crea una funciÃ³n generar_venta() que devuelva un diccionario con los siguientes datos aleatorios:

* "fecha": elegida al azar del rango "2025-01-01" a "2025-01-31"

* "producto": elegido al azar de listaInsumosProductos

* "sucursal": elegida al azar de listaInsumoSucursales

* "cajero": elegido al azar de listaInsumosCajeros

Ejemplo de salida:

```python
{
  "fecha": "2025-01-14",
  "producto": "LÃ¡pices",
  "sucursal": "Coapa",
  "cajero": "SofÃ­a HernÃ¡ndez GarcÃ­a"
}
```

````python
##########
# Solucion
##########

import pandas as pd
import random as r

# Coloca aca tu solucion ðŸ‘‡
def generar_venta():

  fechas = pd.date_range("2025-01-01", "2025-01-31", freq="D")
  rangosFechaString = []

  for fecha in fechas:
    fechaString = fecha.strftime("%Y-%m-%d")
# Agregamos los strings fecha a la lista vacÃ­a
    rangosFechaString.append(fechaString)
  fecha = r.choice(rangosFechaString)

  # Zona de listas con informaciÃ³n
  listaInsumosProductos = ["Libreta", "LÃ¡piz", "Tijeras", "Borrador", "Hojas", "Resistol", "Plumas", "Folder"]

  listaInsumoSucursales = ["Coapa", "Toluca", "Pachuca", "Monterrey", "QuerÃ©taro", "San Luis PotosÃ­", "Guadalajara"]

  listaInsumosCajeros = ["Juan Reyes", "Pau VÃ¡zquez", "Oscar Ramirez", "Miguel GarcÃ­a"]

  cajero = r.choice(listaInsumosCajeros)
  producto = r.choice(listaInsumosProductos)
  sucursal = r.choice(listaInsumoSucursales)

# Definicion del diccionario
  dicc_ventas = {
    "Fecha": fecha,
    "Producto": producto,
    "Sucursal": sucursal,
    "Cajero": cajero,
    }

  return dicc_ventas

print(generar_venta())
````

**3. Verificar rango de fecha**

Pide al usuario una fecha (como texto "YYYY-MM-DD") y verifica:

* Si estÃ¡ dentro del rango "2025-01-01" a "2025-01-31", imprime â€œFecha vÃ¡lida del mesâ€

* Si no, imprime â€œFecha fuera del rango permitidoâ€

**Pista**

Convierte las cadenas a objetos datetime con pd.to_datetime() antes de comparar.

````python
##########
# Solucion
##########


# Coloca aca tu solucion ðŸ‘‡

# Pedimos la fecha al usuario
fecha_usuario = input("Ingresa una fecha (YYYY-MM-DD): ")

# Convertir a objeto fecha
objetoFecha = pd.to_datetime(fecha_usuario)

# Definimos rango
fecha_inicio = pd.to_datetime("2025-01-01")
fecha_fin = pd.to_datetime("2025-01-31")

# Verificar si la fecha estÃ¡ dentro del rango
if fecha_inicio <= objetoFecha <= fecha_fin:
    print("Fecha vÃ¡lida del mes")
else:
    print("Fecha fuera del rango permitido")
````
